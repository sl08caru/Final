## Problem 2. When does the 'raptor catch up to you?

Use your code to figure out when the raptor catches up to you. Both how much time has passed and how far *you* have run.  

Do not just get this information off the graph. Use your Python code to figure this out! Getting the information just by looking at the plot will result in a 0 for this question. You can use an analytic approach (algebra) to check your answer, but that is not the approach I want you to use. Depending on how you write your code, you might not get an *exact* answer, but you should find a way to get pretty close to the algebraic solution. 

```python
# Your code here.

ph = 30
pv = 0
time = 0


while ph != pv: #increments of 1 seconds
    ph = ph + 3
    pv = pv + 18
    time = time + 1
    distance = ph - 30
    
print "%d seconds passed and I ran %d meters before the velociraptor caught up to me." %(time,distance)
```

## Problem 3. When is it close enough to strike?

It turns out the 'raptor will start trying to attack when it is 1 meter away. How much time has passed and how far have you run when it is 1 meter behind you? Use a similar approach as the last problem.

```python
# Your code here. 
ph = 30.0
pv = 0.0
time = 0.0



while ph-pv>1: #increments of .000001 seconds
    ph = ph + .000003
    pv = pv + .000018
    time = time + .000001
    distance = ph - 30.0
    
print "%f seconds passed and I ran %f meters before the velociraptor was about one meter behind me." %(time,distance)

import matplotlib.pylab as plt
%matplotlib inline

t = plt.linspace(0,2,210) 
v = 18*t
h = 3*t + 30
plt.errorbar(t,v,fmt='-',markersize=10,label='Velociraptor Position') 
plt.errorbar(t,h,fmt='-',markersize=10,label='Human Position') 
plt.xlabel('Time',fontsize=18)
plt.ylabel('Position',fontsize=18)
plt.title("Human vs. Velociraptor")
plt.legend(loc=4)
plt.annotate('1 Meter Away', xy=(time,pv), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

```

## Problem 4. Will it bite you?

When the 'raptor is 1 meter away, it will try to bite you. If it doesn't bite you, it will try a second and (if necessary) a third time. If it doesn't get you on the third try, it will grow frustrated and stop chasing you. Hurray!

If it bites you once, you're 'raptor food. If it misses all three times, you get away!

What is the probability that you will get away?


```python
# Your code here
import numpy as np

def probability():
    x = np.random.randint(1,101)
    y = np.random.randint(1,101)
    z = np.random.randint(1,101)
    if x == 1 or x == 2 or x == 3 or x == 4 or x == 5 or x == 6 or x == 7 or x == 8 or x == 9 or x == 10 or x == 11 or x == 12 or x == 13 or x == 14 or x == 15 or x == 16 or x == 17 or x == 18 or x == 19 or x == 20:
        return "You're raptor food."
    
    elif y == 1 or y == 2 or y == 3 or y == 4 or y == 5 or y == 6 or y == 7 or y == 8 or y == 9 or y == 10 or y == 11 or y == 12 or y == 13 or y == 14 or y == 15:
        return "You're raptor food."
    
    elif z == 1 or z == 2 or z == 3 or z == 4 or z == 5 or z == 6 or z == 7:
        return "You're raptor food."
    
    else:
        return "You get away!"
        

n = 0.0
raptorwin = 0.0
humanwin = 0.0


while n < 10000.0:
    probability()
    if probability() == "You're raptor food.":
        raptorwin = raptorwin + 1.0
    else:
        humanwin = humanwin + 1.0
    
    n = n + 1.0
        
p = humanwin/10000.0
print "The probability that I will get away out of 1000 trials is %f" %(p)
```
